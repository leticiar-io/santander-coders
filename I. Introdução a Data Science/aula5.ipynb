```python
# Analisando dados
import pandas as pd
```

```python
taxis = pd.read_csv("taxis.csv")
taxis.head()
```

```python
taxis["dropoff_borough"].value_counts()
```

```python
# Validando hipoteses 
# Tenho 3 vizinhaças, então vizinhaças em que as casas costumam ser mais caras, oferecem melhores gorjetas? 

# Mean = Média
# Median = Mediana 
taxis.groupby("dropoff_borough")["tip"].agg(['mean', 'median'])

# Resultado = Nossa hipotése foi refurtada, pois segundo dados, Manhattan é a vizinhaça mais cara, porém com tem a 2° maior média de gorjeta
```
