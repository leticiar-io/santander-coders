```python
# Fazer uma analise de dados com algum assunto que você gosta

import pandas as pd
```

```python
kpop = pd.read_csv('kpop_data.csv')
kpop.head()
```

```python
# Hipotéses 

# 1. Quanto maior o número de integrantes de um grupo, maior o número de visualizações no youtube
# 2. Quanto maior o menor o número do ano do debut, maior são as vendas anuais
```

```python
# Testanto a primeira hipótese
kpop.groupby("Members")["YouTubeMetrics"].agg(['mean', 'median'])

# Resultado hipotese: grupos entre 6 a 7 integrantes tem uma média de visualizações maior.
```

```python
# Testando a segunda hipótese
kpop.groupby("DebutYear")["AnnualSales"].agg(['mean', 'median'])

# Resultado da 2 hipotese: Grupos mais recentes tem vendas maiores do que os mais antigos
```
