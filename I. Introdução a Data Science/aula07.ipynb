```python
# Criando gr√°ficos

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('desafios-propostos/kpop_data.csv')
#df = df.drop(df.columns[0], axis = 1)
df.head()
```

```python
eixo_x = df["Active"].value_counts().sort_index().keys()
eixo_y = df["Active"].value_counts().sort_index().values

plt.title("Qtd de grupos ativos")
plt.xlabel("Status")
plt.ylabel("Qtd de grupos ativos")

plt.plot(eixo_x, eixo_y)
```

```python
eixo_x = df['DebutYear'].value_counts().keys()
eixo_y = df['DebutYear'].value_counts().values

plt.title("Qtd de Debuts por ano")
plt.xlabel("Ano")
plt.ylabel("Qtd de Debuts")

plt.bar(eixo_x, eixo_y)
plt.show()
```
