```python
# Adicionando livros na lista de livros

# Importando a biblioteca
import pandas as pd

# Criando um dicionário com os dados
dados = {
   "Autores": ["Machado de Assis", "José de Alencar", "Cecília Meireles", "Carlos Drummond de Andrade"],
   "Livros": ["Dom Casmurro", "Iracema", "Romanceiro da Inconfidência", "A Rosa do Povo"],
   "Ano de Publicação": [1899, 1865, 1953, 1945]
}

# Criando um DataFrame
df = pd.DataFrame(dados)

df
```

```python
# Adicionando novo livro

# Forma 1
df.loc[4] = ["Fernando Pessoa", "Mensagem", 1934]
df
```

```python
# Forma 2 
novo_livro = {
   "Autores": ["Cecília Meireles"],
   "Livros": ["A Hora da Estrela"],
   "Ano de Publicação": [1977]
}

# Adicionando uma nova linha 
df1 = pd.DataFrame(novo_livro)
df_agrupado = pd.concat([df, df1])
df_agrupado
```
